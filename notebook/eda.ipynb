```python
import pandas as pd
```

```python
def load_data(path):
    df = pd.read_csv(path)
    print("Cleaned data loaded")
    return df
```

```python
def fraud_distribution(df):
    print("\nFraud vs Normal Transactions:")
    print(df['Class'].value_counts())
```

```python
def avg_transaction(df):
    print("\nAverage transaction amount:")
    print(df.groupby('Class')['Amount'].mean())
```

```python
def fraud_by_hour(df):
    result = df.groupby('Hour')['Class'].sum().sort_values(ascending=False)
    print("\nFraud transactions by Hour:")
    print(result.head())
```

```python
def high_value_fraud(df):
    print("\nTop 5 High Value Fraud Transactions:")
    print(df[df['Class'] == 1].sort_values(by='Amount', ascending=False).head())
```

```python
if __name__ == "_main_":
    file_path = "../outputs/cleaned_fraud_data.csv"

    df = load_data(file_path)
    fraud_distribution(df)
    avg_transaction(df)
    fraud_by_hour(df)
    high_value_fraud(df)
```
